```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
iris['sepal_sum'] = iris ['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
iris['petal_sum'] = iris ['petal_length'] + iris['petal_width']
iris.head()
```

```python
iris = iris[['sepal_length', 'sepal_width', 'sepal_sum', 'petal_length', 'petal_width', 'petal_sum']]
iris
```

```python
mean = [np.mean(iris['sepal_length']),
        np.mean(iris['sepal_width']),
        np.mean(iris['sepal_sum']),
        np.mean(iris['petal_length']),
        np.mean(iris['petal_width']),
        np.mean(iris['petal_sum'])]

mean= list(iris.mean())

iris.std()
std =list(iris.std())

iris.min()
min =list(iris.min())

iris.max()
max =list(iris.max())

summary_dict = {'mean': mean, 'std': std , 'min' : min , 'max' : max }

summary_df = pd.DataFrame(data = summary_dict)

summary_df
```

```python
summary_df.index=['sepal_length', 'sepal_width', 'sepal_sum', 'petal_length', 'petal_width', 'petal_sum']
summary_df
```
